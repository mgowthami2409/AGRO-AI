```python
import cv2
import numpy as np
import os
import tensorflow as tf
import albumentations as A  # For data augmentation
from tqdm import tqdm  # For progress tracking

# Define paths
image_folder = r"C:\Users\Admin\Downloads\Major_Project_New\paddy_leaf_images"  # Change this to your actual image folder
output_folder = r"C:\Users\Admin\Downloads\Major_Project_New\preprocessed_images"
os.makedirs(output_folder, exist_ok=True)

# Define image augmentation pipeline
augmentation = A.Compose([
    A.HorizontalFlip(p=0.5),  # Random horizontal flipping
    A.Rotate(limit=20, p=0.5),  # Random rotation within Â±20 degrees
    A.RandomBrightnessContrast(p=0.5),  # Random brightness/contrast adjustment
])

# Function to preprocess an image
def preprocess_image(image_path, output_folder):
    # Read image
    image = cv2.imread(image_path)
    
    # Resize to 128x128 pixels
    image = cv2.resize(image, (180, 180))
    
    # Apply data augmentation
    augmented = augmentation(image=image)
    image = augmented["image"]
    
    # Normalize pixel values to range 0-1
    image = image.astype(np.float32) / 255.0

    # Save preprocessed image
    output_path = os.path.join(output_folder, os.path.basename(image_path))
    cv2.imwrite(output_path, (image * 255).astype(np.uint8))  # Convert back to 0-255 before saving
    
    return output_path

# Process all images in the dataset
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(".jpg") or f.endswith(".png")]

preprocessed_paths = []
for img_path in tqdm(image_files, desc="Preprocessing Images"):
    processed_path = preprocess_image(img_path, output_folder)
    preprocessed_paths.append(processed_path)

print(f" Preprocessed {len(preprocessed_paths)} images. Saved in '{output_folder}'")
```

```python
import cv2
import numpy as np
import os
import pandas as pd
from tqdm import tqdm  # For progress tracking

# Define preprocessed image folder
preprocessed_folder = r"C:\Users\Admin\Downloads\Major_Project_New\preprocessed_images" # Change this to your actual folder path
output_csv = "vegetation_indices.csv"

# Function to compute vegetation indices
def compute_indices(image_path):
    image = cv2.imread(image_path)
    
    # Convert image to float for calculations
    image = image.astype(np.float32) / 255.0  
    
    # Extract R, G, B channels
    B, G, R = cv2.split(image)
    
    # Compute vegetation indices
    VARI = (G - R) / (G + R - B + 1e-6)  # Adding small value to avoid division by zero
    GLI = (2 * G - R - B) / (2 * G + R + B + 1e-6)
    ExG = (2 * G - R - B)
    
    # Compute mean values
    return np.mean(VARI), np.mean(GLI), np.mean(ExG)

# Process all images and store results
data = []
image_files = [f for f in os.listdir(preprocessed_folder) if f.endswith(".jpg") or f.endswith(".png")]

for img_name in tqdm(image_files, desc="Computing Vegetation Indices"):
    img_path = os.path.join(preprocessed_folder, img_name)
    VARI, GLI, ExG = compute_indices(img_path)
    data.append((img_name, VARI, GLI, ExG))

# Convert results to a DataFrame
df = pd.DataFrame(data, columns=["Image_Path", "VARI", "GLI", "ExG"])

# Save to CSV file
df.to_csv(output_csv, index=False)
print(f" Vegetation indices saved to '{output_csv}'")
```

```python
import torch
import torchvision.transforms as transforms
import cv2
import numpy as np
import os
import pandas as pd
from tqdm import tqdm  # For progress tracking

# Define paths
preprocessed_folder = r"C:\Users\Admin\Downloads\Major_Project_New\preprocessed_images"  # Folder containing preprocessed images
indices_file = r"C:\Users\Admin\Downloads\Major_Project_New\vegetation_indices.csv"  # CSV file with vegetation indices
output_csv = "image_features.csv"  # Output file with extracted features

# Load ResNet34 Feature Extractor (Ensure Step 2.1 is completed)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
resnet34 = resnet34.to(device)  # Move model to GPU if available

# Define image transformations for ResNet34
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((180, 180)),
    transforms.ToTensor()
])

# Function to extract features using ResNet34
def extract_features(image_path):
    image = cv2.imread(image_path)
    image = transform(image).unsqueeze(0).to(device)  # Add batch dimension and move to GPU if available
    
    with torch.no_grad():  # No gradient calculation needed for inference
        features = resnet34(image).flatten().cpu().numpy()  # Convert to 1D feature vector
    
    return features

# Load vegetation indices
indices_df = pd.read_csv(indices_file)

# Extract features and store them with vegetation indices
feature_data = []
for _, row in tqdm(indices_df.iterrows(), total=len(indices_df), desc="Extracting Features"):
    image_path = os.path.join(preprocessed_folder, row["Image_Path"])
    
    if os.path.exists(image_path):  # Check if image exists
        features = extract_features(image_path)
        feature_data.append([image_path] + features.tolist() + [row["VARI"], row["GLI"], row["ExG"]])
    else:
        print(f"âš ï¸ Warning: Image {image_path} not found!")

# Convert to DataFrame
feature_columns = [f"Feature_{i}" for i in range(512)]  # 512 feature dimensions from ResNet34
df = pd.DataFrame(feature_data, columns=["Image_Path"] + feature_columns + ["VARI", "GLI", "ExG"])

# Save extracted features to CSV
df.to_csv(output_csv, index=False)
print(f"âœ… Feature extraction completed! Data saved in '{output_csv}'")
```

```python
import pandas as pd
import numpy as np
import os
import re
from tqdm import tqdm

# Load extracted feature data
features_file = r"C:\Users\Admin\Downloads\Major_Project_New\image_features.csv"  # File containing extracted ResNet34 features + vegetation indices
output_sequence_file = "time_series_data.csv"

# Define sequence length
SEQUENCE_LENGTH = 5  # Number of images per sequence

# Load feature dataset
df = pd.read_csv(features_file)

# âœ… FIX: Extract timestamps correctly, handling filenames without numeric timestamps
def extract_timestamp(filename):
    match = re.search(r'(\d+)', filename)  # Extract numeric part
    return int(match.group(1)) if match else 0  # Default to 0 if no timestamp found

df["Timestamp"] = df["Image_Path"].apply(lambda x: extract_timestamp(os.path.basename(x)))  # Extract timestamp

# âœ… Sort images based on extracted timestamps
df.sort_values(by=["Image_Path", "Timestamp"], inplace=True)

# Group images into time-series sequences
sequences = []
labels = []

# âœ… Extract only feature columns (not Image_Path, VARI, GLI, ExG, or NPK values)
feature_columns = [col for col in df.columns if col.startswith("Feature_")]
vegetation_columns = ["VARI", "GLI", "ExG"]
label_columns = ["Nitrogen (N)", "Phosphorus (P)", "Potassium (K)"]

expected_feature_size = len(feature_columns) * SEQUENCE_LENGTH  # Should be 512 * 5

# âœ… Iterate through unique leaf IDs
leaf_ids = df["Image_Path"].apply(lambda x: x.split("_")[0]).unique()

for leaf_id in tqdm(leaf_ids, desc="Organizing Time-Series Data"):
    leaf_df = df[df["Image_Path"].str.startswith(leaf_id)].reset_index(drop=True)

    if len(leaf_df) < SEQUENCE_LENGTH:
        continue  # Skip if not enough images

    for i in range(len(leaf_df) - SEQUENCE_LENGTH + 1):
        # Extract only feature columns for SEQUENCE_LENGTH images
        sequence_features = leaf_df.iloc[i:i + SEQUENCE_LENGTH][feature_columns].values.flatten()

        if len(sequence_features) != expected_feature_size:
            print(f"âš ï¸ Mismatch in feature size for leaf {leaf_id}. Expected {expected_feature_size}, got {len(sequence_features)}")
            continue  # Skip problematic sequences

        vegetation_indices = leaf_df.iloc[i + SEQUENCE_LENGTH - 1][vegetation_columns].values  # Last image's VARI, GLI, ExG
        npk_values = leaf_df.iloc[i + SEQUENCE_LENGTH - 1][label_columns].values  # Last image's NPK values

        sequences.append(np.hstack([sequence_features, vegetation_indices]))
        labels.append(npk_values)  # Store NPK values as labels

# âœ… Fix column count issue
sequence_columns = [f"Feature_{i}" for i in range(expected_feature_size)] + vegetation_columns
label_columns = ["Nitrogen (N)", "Phosphorus (P)", "Potassium (K)"]

df_sequences = pd.DataFrame(sequences, columns=sequence_columns)
df_labels = pd.DataFrame(labels, columns=label_columns)

# âœ… Save structured dataset
df_final = pd.concat([df_sequences, df_labels], axis=1)
df_final.to_csv(output_sequence_file, index=False)

print(f"âœ… Time-series dataset saved in '{output_sequence_file}'")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Load the structured time-series dataset
dataset_file = r"C:\Users\Admin\Downloads\Major_Project_New\time_series_data.csv"  # Update your file path
df = pd.read_csv(dataset_file)

# Inspect the dataset
print(df.head())
print(df.columns)

# Set FEATURE_DIM manually based on dataset structure
SEQUENCE_LENGTH = 6  # Number of images per sequence
FEATURE_DIM = 512  # Adjust this if necessary based on your dataset

# Define feature & label columns based on the actual dataset
feature_columns = [f"Feature_{i}" for i in range(1, FEATURE_DIM + 1)] + ["VARI", "GLI", "ExG"]
label_columns = ["Nitrogen (N)", "Phosphorus (P)", "Potassium (K)"]

# Extract features and labels
X = df[feature_columns].values
Y = df[label_columns].values  # Target NPK values

# Debugging: Check feature shape before reshaping
print(f"ğŸ” X.shape before reshaping: {X.shape}")
print(f"ğŸ“Œ Using FEATURE_DIM: {FEATURE_DIM}")
print(f"ğŸ“Œ Expected reshape dimensions: (-1, {SEQUENCE_LENGTH}, {FEATURE_DIM + 3})")

# Ensure you only use samples that can form complete sequences
num_samples = 30  # Use only the first 30 samples
X = X[:num_samples]  # Use only the first 30 samples
Y = Y[:num_samples]  # Corresponding labels

# Ensure reshaping works
expected_size = SEQUENCE_LENGTH * (FEATURE_DIM + 3)
if X.shape[0] % SEQUENCE_LENGTH != 0:
    print("âš ï¸ ERROR: Reshaping will still fail, check dataset structure!")
    raise ValueError("Dataset dimensions are incorrect, please verify `FEATURE_DIM` and column selections.")
else:
    print("âœ… Reshaping should work!")

# Apply corrected reshape
X = X.reshape(-1, SEQUENCE_LENGTH, FEATURE_DIM + 3)  # Including vegetation indices

# Check the shape of X after reshaping
print(f"ğŸ” X.shape after reshaping: {X.shape}")

# Split dataset into training (80%) and validation (20%)
X_train, X_val, Y_train, Y_val = train_test_split(X, Y[:X.shape[0]], test_size=0.2, random_state=42)

# Define BiGRU Model
def build_biGRU_model(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Bidirectional(tf.keras.layers.GRU(128, return_sequences=True, dropout=0.2), input_shape=input_shape),
        tf.keras.layers.Bidirectional(tf.keras.layers.GRU(64, return_sequences=False, dropout=0.2)),  
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(3, activation='linear')  # Predicts N, P, K values
    ])
    
    model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['mae', 'accuracy'])
    return model

# Load the BiGRU model
model = build_biGRU_model((SEQUENCE_LENGTH, FEATURE_DIM + 3))

# Define callbacks for early stopping & best model saving
callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True),
    ModelCheckpoint("best_biGRU_model.keras", save_best_only=True)  # Change to .keras
]

# Train the model
history = model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=50,  # Set to 30 epochs
    batch_size=16,
    callbacks=callbacks,
    verbose=1  # Set verbosity to 1 for detailed output
)

# Save the trained model
model.save("final_biGRU_model.keras")  # Change to .keras
print("âœ… Model training complete! Best model saved as 'best_biGRU_model.keras'")
```

```python
import streamlit as st
import numpy as np
import pandas as pd
import cv2
from tensorflow.keras.models import load_model

# Load Models
npk_model = load_model("best_biGRU_model.keras")
disease_model = load_model("riceleaf_newmodel.h5")

# Constants
SEQUENCE_LENGTH = 5  # Example sequence length
FEATURE_DIM = 515  # Example feature dimension (512 ResNet + 3 indices)

# Function to Compute Vegetation Indices
def compute_indices(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = image.astype(np.float32) / 255.0

    R = image[:, :, 0]
    G = image[:, :, 1]
    B = image[:, :, 2]

    VARI = (G - R) / (G + R - B)
    GLI = (2 * G - R - B) / (2 * G + R + B)
    ExG = 2 * G - R - B

    return np.array([VARI.mean(), GLI.mean(), ExG.mean()])

# Function to Predict NPK Values
def predict_npk(image):
    indices = compute_indices(image)
    resnet_features = np.random.rand(512)  # Replace with actual ResNet feature extraction
    features = np.concatenate([resnet_features, indices])
    features = np.tile(features, (SEQUENCE_LENGTH, 1))  # Repeat to match sequence length
    features = features.reshape(1, SEQUENCE_LENGTH, FEATURE_DIM)
    npk_values = npk_model.predict(features)
    return npk_values.flatten()

# Function to Predict Disease
def predict_disease(image):
    image_resized = cv2.resize(image, (180, 180)) / 255.0
    image_resized = np.expand_dims(image_resized, axis=0)
    disease_pred = disease_model.predict(image_resized)
    disease_label = np.argmax(disease_pred)
    
    disease_dict = {
        0: "Bacterial Leaf Blight",
        1: "Brown Spot",
        2: "Smut",
        3: "Blast",
        4: "Tungro"
    }
    
    if np.max(disease_pred) < 0.5:  # Adjust threshold as needed
        return "Healthy"
    
    return disease_dict[disease_label]

# Function to Suggest Fertilizers
def suggest_fertilizer(n, p, k):
    recommendations = {
        "Nitrogen": {
            "Chemical": "Urea (46% N) or Ammonium Nitrate",
            "Organic": "Compost, Vermicompost, or Cow Manure"
        },
        "Phosphorus": {
            "Chemical": "Single Super Phosphate (SSP) or DAP",
            "Organic": "Bone Meal, Rock Phosphate"
        },
        "Potassium": {
            "Chemical": "Muriate of Potash (MOP) or Sulfate of Potash",
            "Organic": "Wood Ash, Banana Peel Fertilizer"
        }
    }

    suggestions = {}

    if n < 20:
        suggestions["Nitrogen"] = recommendations["Nitrogen"]
    if p < 30:
        suggestions["Phosphorus"] = recommendations["Phosphorus"]
    if k < 40:
        suggestions["Potassium"] = recommendations["Potassium"]

    return suggestions

# Streamlit UI
st.title("Paddy Leaf Analysis for NPK Prediction & Disease Detection")

uploaded_file = st.file_uploader("Upload a Paddy Leaf Image", type=["jpg", "png", "jpeg"])

if uploaded_file is not None:
    file_bytes = np.asarray(bytearray(uploaded_file.read()), dtype=np.uint8)
    image = cv2.imdecode(file_bytes, 1)

    st.image(image, caption="Uploaded Image", use_container_width=True)

    if st.button("Predict"):
        # Predictions
        n_pred, p_pred, k_pred = predict_npk(image)
        disease = predict_disease(image)
        recommendations = suggest_fertilizer(n_pred, p_pred, k_pred)

        # Display Results
        st.subheader("Predicted NPK Values")
        st.write(f"**Nitrogen (N):** {n_pred:.2f}")
        st.write(f"**Phosphorus (P):** {p_pred:.2f}")
        st.write(f"**Potassium (K):** {k_pred:.2f}")

        st.subheader("Fertilizer Recommendations")
        for nutrient, advice in recommendations.items():
            st.write(f"{nutrient}:")
            st.write(f"  ğŸ”¹ Chemical Fertilizer: {advice['Chemical']}")
            st.write(f"  ğŸ”¹ Organic Fertilizer: {advice['Organic']}")

        st.subheader("Disease Detection Result")
        st.write(f"**Detected Disease:** {disease}")

        # Save to CSV
        df = pd.DataFrame([{
            "N": n_pred, "P": p_pred, "K": k_pred,
            "Nitrogen_Chemical": recommendations.get("Nitrogen", {}).get("Chemical", ""),
            "Nitrogen_Organic": recommendations.get("Nitrogen", {}).get("Organic", ""),
            "Phosphorus_Chemical": recommendations.get("Phosphorus", {}).get("Chemical", ""),
            "Phosphorus_Organic": recommendations.get("Phosphorus", {}).get("Organic", ""),
            "Potassium_Chemical": recommendations.get("Potassium", {}).get("Chemical", ""),
            "Potassium_Organic": recommendations.get("Potassium", {}).get("Organic", "")
        }])

        df.to_csv("fertilizer_recommendations.csv", mode='a', index=False, header=False)
        st.write("âœ… Recommendations saved to CSV.")
```
